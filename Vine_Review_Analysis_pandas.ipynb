```python
import pandas as pd 
```

```python
df = pd.read_csv("./database/vine_table.csv")
```

```python
df
```

```python
top_20_vine_df = df[df['total_votes'] >= 20]
top_20_vine_df
```

```python
updated_df = top_20_vine_df[(top_20_vine_df['helpful_votes']/top_20_vine_df['total_votes']) >= 0.5]
updated_df
```

```python
vine_paid_df = updated_df[updated_df['vine'] == 'Y']
vine_paid_df
```

```python
vine_unpaid_df = updated_df[updated_df['vine'] == 'N']
vine_unpaid_df
```

## 5 star reviews among vine reviews

```python
# total number of star ratings from the DataFrame that vine=Y.
total_paid_count = vine_paid_df['review_id'].count()


# the number of 5 star ratings from the Dataframe 
paid_5star_rating = vine_paid_df[vine_paid_df['star_rating']==5]['star_rating'].count()


# calculate percent 5 star rating 
paid_percent_5star_rating = paid_5star_rating/total_paid_count * 100

#results
print("Total paid reviews",total_paid_count)
print("Count of 5 star paid reviews ", paid_5star_rating)
print("Percent of 5 star rating",paid_percent_5star_rating)
```

```python
# total number of star ratings from the DataFrame that vine=Y.
total_unpaid_count = vine_unpaid_df['review_id'].count()


# the number of 5 star ratings from the Dataframe 
unpaid_5star_rating = vine_unpaid_df[vine_unpaid_df['star_rating']==5]['star_rating'].count()


# calculate percent 5 star rating 
unpaid_percent_5star_rating = unpaid_5star_rating/total_unpaid_count * 100

#results
print("Total paid reviews",total_unpaid_count)
print("Count of 5 star paid reviews ", unpaid_5star_rating)
print("Percent of 5 star rating",unpaid_percent_5star_rating)
```
